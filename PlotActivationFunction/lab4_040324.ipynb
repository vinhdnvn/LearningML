```python
import numpy as np
import matplotlib.pyplot as plt
```

#Hàm tanh

##Plot hàm tanh thủ công


```python
# hàm tanh
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
# Tạo một numpy array chứa các giá trị x
x = np.linspace(-10, 10, 1000)

# Tính giá trị y = tanh(x) cho mỗi giá trị x
y = tanh(x)

# Vẽ đồ thị
plt.figure(figsize=(12, 6))
plt.plot(x, y)
plt.axhline(y=0, color='red', linestyle='--')
plt.axvline(x=0, color='gray', linestyle='--')
plt.xlabel('X')
plt.ylabel('tanh(X)')
plt.title('Đồ thị hàm tanh')
plt.show()
```

##plot hàm tanh dùng hàm

```python
x = np.linspace(-5, 5, 1000)

#dùng hàm có sẵn từ thư viện numpy
y = np.tanh(x)


plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.axhline(y=0, color='gray', linestyle='--')
plt.axvline(x=0, color='gray', linestyle='--')
plt.xlabel('X')
plt.ylabel('tanh(X)')
plt.title('Đồ thị hàm tanh')
plt.show()
```

# Hàm Sigmoid

## Thủ công

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 1000)


y = sigmoid(x)

# Vẽ đồ thị
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.axhline(y=0, color='gray', linestyle='--')
plt.axvline(x=0, color='gray', linestyle='--')
plt.xlabel('X')
plt.ylabel('sigmoid(X)')
plt.title('Đồ thị hàm sigmoid (thủ công)')
plt.show()
```

##Đạo hàm của hàm sigmooid

```python
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
y_daoham = sigmoid_derivative(x)



# Vẽ đồ thị
plt.figure(figsize=(18, 6))
plt.plot(x, y, label='sigmoid(x)')
plt.plot(x, y_daoham, label="sigmoid'(x)")
plt.axhline(y=0, color='gray', linestyle='--')
plt.axvline(x=0, color='gray', linestyle='--')
plt.xlabel('X')
plt.ylabel('Giá trị')
plt.title('Đồ thị hàm sigmoid và đạo hàm ')
plt.legend()
plt.show()
```

#Hàm ReLU

```python
def relu(x):
    return np.maximum(0, x)


def relu_derivative(x):
    x_copy = x.copy()
    x_copy[x_copy <= 0] = 0
    x_copy[x_copy > 0] = 1
    return x_copy


x = np.linspace(-5, 5, 1000)


y = relu(x)
y_daoham = relu_derivative(x)

# Vẽ đồ thị
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='ReLU(x)')
plt.plot(x, y_daoham, label="ReLU'(x)")
plt.axhline(y=0, color='gray', linestyle='--')
plt.axvline(x=0, color='gray', linestyle='--')
plt.xlabel('X')
plt.ylabel('Giá trị')
plt.title('Đồ thị hàm ReLU và đạo hàm của nó')
plt.legend()
plt.show()
```
